```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

```

```python
# Create a dummy dataset
x = np.linspace(0, 10, 100)
y = np.sin(x) + 0.1 * np.random.randn(100)

# Plot the dataset
plt.plot(x, y, 'o')
plt.show()
```

```python
# Create cosine dataset
x = np.linspace(0, 10, 100)
y = np.cos(x) + 0.1 * np.random.randn(100)

# Plot the dataset
plt.plot(x, y, 'o')
plt.show()
```

```python
# Plot random points
x = np.random.randn(100)
y = np.random.randn(100)

# Plot the dataset
plt.plot(x, y, 'o')
plt.show()
```
